5. Se utiliza int(input(...)) para convertir la cantidad ingresada por el usuario a un entero, y float(input(...)) para convertir el precio a un n煤mero decimal.
6. En la l铆nea factura += "{:<10}{:<10}{:<10}{:<10}\n".format("Producto", "Cantidad", "Precio", "Total"), se utiliza el m茅todo format() para formatear el encabezado de la factura. La especificaci贸n {:<10} indica que el valor se alinear谩 a la izquierda en un campo de 10 caracteres.
7. Se itera sobre cada producto en la lista productos. Para cada producto, se calcula su total (total_producto = producto["cantidad"] * producto["precio"]), se agrega este total a la lista totales_productos, y se agrega una l铆nea a la factura utilizando nuevamente el m茅todo format().

10. Se agrega la l铆nea del total a la factura utilizando una f-string: factura += f"Total a pagar:\t\t\t{total_factura:.2f}\n". La especificaci贸n {total_factura:.2f} formatea el n煤mero con dos decimales.
11. Se imprime la factura resultante.

```python
# Entrada de datos
productos = []
while True:
    nombre = input("Ingrese el nombre del producto (o 'x' para salir): ")
    if nombre.lower() == 'x':
        break
    cantidad = int(input("Ingrese la cantidad: "))
    precio = float(input("Ingrese el precio: "))
    productos.append({"nombre": nombre, "cantidad": cantidad, "precio": precio})

# Generar la factura
factura = "FACTURA\n\n"
factura += "{:<10}{:<10}{:<10}{:<10}\n".format("Producto", "Cantidad", "Precio", "Total")
factura += "-" * 40 + "\n"

totales_productos = []
for producto in productos:
    total_producto = producto["cantidad"] * producto["precio"]
    totales_productos.append(total_producto)
    factura += "{:<10}{:<10}{:<10}{:<10}\n".format(producto["nombre"], producto["cantidad"], producto["precio"], total_producto)

factura += "-" * 40 + "\n"
total_factura = sum(totales_productos)
factura += f"Total a pagar:\t\t\t{total_factura:.2f}\n"

print(factura)
```

- Se define una cadena separador con 40 guiones ("-") que se utilizar谩 como l铆nea separadora en la factura.
En la l铆nea factura += separador + "\n", se agrega la cadena separador seguida de un salto de l铆nea para separar el encabezado de la factura del cuerpo.
- Despu茅s de iterar sobre los productos y agregar sus l铆neas correspondientes a la factura, se agrega nuevamente la cadena separador seguida de un salto de l铆nea.
- El resto del c贸digo es similar al ejemplo anterior, utilizando int() y float() para convertir los valores num茅ricos, y format() y f-strings para formatear la salida.

Este c贸digo solicita al usuario ingresar los detalles de los productos (nombre, cantidad y precio), almacena estos datos en una lista de diccionarios, calcula el total de cada producto y el total de la factura, y luego genera una factura formateada que muestra los detalles de los productos y el total a pagar.

```python
# Entrada de datos
productos = []
while True:
    nombre = input("Ingrese el nombre del producto (o 'x' para salir): ")
    if nombre.lower() == 'x':
        break
    cantidad = int(input("Ingrese la cantidad: "))
    precio = float(input("Ingrese el precio: "))
    productos.append({"nombre": nombre, "cantidad": cantidad, "precio": precio})

# L铆nea separadora
separador = "----------------------------------------"

# Generar la factura
factura = "FACTURA\n\n"
factura += "{:<10}{:<10}{:<10}{:<10}\n".format("Producto", "Cantidad", "Precio", "Total")
factura += separador + "\n"

totales_productos = []
for producto in productos:
    total_producto = producto["cantidad"] * producto["precio"]
    totales_productos.append(total_producto)
    factura += "{:<10}{:<10}{:<10}{:<10}\n".format(producto["nombre"], producto["cantidad"], producto["precio"], total_producto)

factura += separador + "\n"
total_factura = sum(totales_productos)
factura += f"Total a pagar:\t\t\t{total_factura:.2f}\n"

print(factura)
```

```python
import json

# Leemos el archivo JSON
with open('calificaciones.json', 'r') as archivo_calificaciones:
    estudiantes = json.load(archivo_calificaciones)

    suma_total = 0
    contador = 0
    for estudiante in estudiantes:
        suma_total += estudiante['calificacion']
        contador += 1

# Calculamos el promedio
promedio = suma_total / contador if contador > 0 else 0

# Escribimos el promedio de calificaciones en un nuevo archivo JSON
with open('promedio_calificaciones.json', 'w') as archivo_promedio:
    json.dump({"promedio_calificaciones": promedio}, archivo_promedio)

print("El promedio es: ", promedio)
```
